```python
from numpy import random
import matplotlib.pyplot as plt
import pandas as pd
```

### Распределение случайных значений из 10 000 вариантов

```python
data_10k = []
for i in range(10000):
    num = []
    x = random.randint(0, 10, 3)
    for i, el in enumerate(x):
        a = str(x[i])
        num.append(a)
    data_10k.append(''.join(num))
```

```python
plt.hist(data_10k)
```

# Распределения игр, чисел с тремя и двумя одинаковыми цифрами, и с одной

Стартовый банк равен 10 000 У.Е.

```python
data = pd.read_csv('data.txt', names=['var', 'numb'])
data
```

```python
dcount = data[(data['var'] == 'count')]
```

```python
dcount['numb'].plot(kind='hist', grid=True, title='Количество игр')
```

```python
dcount['numb'].plot(kind='box', grid=True, title='Количество игр')
```

```python
dXXX = data[(data['var'] == '777') | (data['var'] == '999') | (data['var'] == '555') | (data['var'] == '333') | (data['var'] == '111')]
```

```python
dXXX['numb'].plot(kind='hist', grid=True, title='Выпадение чисел с тремя одинаковыми цифрами')
```

```python
dXXX['numb'].plot(kind='box', grid=True, title='Выпадение чисел с тремя одинаковыми цифрами')
```

```python
dtwo = data[(data['var'] == '*00')| (data['var'] == '*77')]
```

```python
dtwo['numb'].plot(kind='hist', grid=True, title='Выпадение чисел с двумя одинаковыми цифрами')
```

```python
dtwo['numb'].plot(kind='box', grid=True, title='Выпадение чисел с двумя одинаковыми цифрами')
```

```python
dOne = data[(data['var'] == '**0')| (data['var'] == '**7')]
```

```python
dOne['numb'].plot(kind='hist', grid=True, title='Выпадение чисел с одной цифрой')
```

```python
dOne['numb'].plot(kind='box', grid=True, title='Выпадение чисел с одной цифрой')
```
