```python
def traffic_source(row):
    if row.traffic_source == 'yandex' or row.traffic_source == 'google':
        return 'organic'
        
    elif row.region == 'Russia':
        if row.traffic_source == 'email':
            return  'ad'
        elif row.traffic_source == 'paid':
            return  'ad'
            
    elif row.region != 'Russia':
        if row.traffic_source == 'email':
            return  'other'
        elif row.traffic_source == 'paid':
            return  'other'
```

```python
log['source_type'] = log.apply(traffic_source, axis=1)
log
```

# Задание 3

```python
ratings = pd.read_csv('data/ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_ = ratings.value_counts('userId')
user = pd.DataFrame(ratings_)
user = user.rename(columns={0: 'count'})
user.info()
```

```python
filt_users = user[ (user['count'] > 100) ]
filt_users_new = filt_users.reset_index()
DF = filt_users_new.to_dict('split')
pure_data = DF['data']
```

```python
lifetime_ = []

for user in pure_data:
    df_list = [user[0]]
    df = ratings.loc[ ratings.userId == user[0]]
    df = df.sort_values(by='timestamp', ascending=False)
    diff = df['timestamp'].iloc[0] - df['timestamp'].iloc[-1]
    df_list.append(diff)
    lifetime_.append(df_list)
```

```python
lifetime = pd.DataFrame(lifetime_)
df_lifetime = lifetime.rename(columns={0: 'userId', 1: 'lifetime'})
df_lifetime
```
