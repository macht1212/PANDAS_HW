```python
import pandas as pd
import numpy as np
```

```python
names_col = []
with open('статистика/data.txt', 'r') as file:
    
    for line in file:
        names_col.append(line.strip())
```

```python
dataframe = pd.read_csv('статистика/horse_data.csv', names=names_col)
dataframe
```

```python
data_info_raw = dataframe[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']]
data_info_raw
```

```python
data_info_fill_nan = data_info_raw.replace('?', None)
```

```python
data_info_fill_nan.info()
```

В датафрейме пропущено менее 20% значений, что в принципе не так страшно, при этом также стоит отметить, что строк с пропущенными значениями также не так много, что позволит нам не обращать внимание на пропуски в целом

```python
data_info_float = data_info_fill_nan.astype({'surgery?': np.float64, 'rectal temperature': np.float64, 'pulse': np.float64, 'respiratory rate': np.float64, 'temperature of extremities': np.float64, 'pain': np.float64, 'outcome': np.float64})
```

```python
data_info_float.describe()
```

```python
(data_info_float.isna().mean() * 100).round(2)
```

# Взрослые лошади

Для начала рассмотрим статистику по взрослым лошадям:

## Смертность прооперированных лошадей 

```python
data_info_adult_surg = data_info_float[(data_info_float['Age'] == 1) & (data_info_float['surgery?'] == 1)]
data_info_adult_surg
```

```python
data_info_adult_surg['outcome'].value_counts()
```

Смертность взрослых лошадей после проведения операции составляет практически половину: 45,7%

Сначала уберем выпадающие значения по пульсу

```python
q1 = data_info_adult_surg['pulse'].quantile(0.25)
q3 = data_info_adult_surg['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
data_info_as = data_info_adult_surg[data_info_adult_surg['pulse'].between(lower_bound, upper_bound, inclusive=True)]
data_info_as
```

Теперь отсортируем значения data_info_adult_surg по значению 2 - погибли и затем 3 - эвтаназия
Далее посмотрим статистику по всем остальным значениям таблицы:

### Естесственная смерть

```python
data_info_asd = data_info_as[data_info_as['outcome'] == 2]
data_info_asd['pain'].value_counts()
```

Исходя из описания датафрейма видим, что смертность лошади обуславливается одним из ключевых показателей: температура, пульс, частота дыхания и субъективный показатель - боль. 

Чаще всего лошади умирали испытывая боль (значения 4 - периодическая сильная боль, 5 - постоянная сильная боль) в редких случаях боль была либо 2 - ноющая, либо переодически средняя. 

Пропуски температуры в данном случае вполне рандомны, то есть лошадь могла испытывать сильные боли при низкой температуре и наоборот. Заполнить их можно средним значением по сегменту. 

```python
data_info_asd['rectal temperature'] = data_info_asd['rectal temperature'].fillna(data_info_asd['rectal temperature'].mean())
```

```python
data_info_asd
```

С учетом того, что частота дыхания для лошадей в этой категории выше нормального либо значительно, либо средне, то вполне можно пропуски заменить на модальное значение.  
Точно так же заполним значения по боли (хотя помним, что это субъективный показатель), а также значения статистики, посчитаной ранее

```python
data_info_asd['respiratory rate'] = data_info_asd['respiratory rate'].fillna(data_info_asd['respiratory rate'].mode()[0])
data_info_asd['pain'] = data_info_asd['pain'].fillna(data_info_asd['pain'].mode()[0])
data_info_asd
```

С пульсом все сложнее, поэтому решил заменить на медиану
Экстримы температуры заполнил медианой

```python
data_info_asd['pulse'] = data_info_asd['pulse'].fillna(data_info_asd['pulse'].median())
data_info_asd['temperature of extremities'] = data_info_asd['temperature of extremities'].fillna(data_info_asd['temperature of extremities'].median())
data_info_asd
```

```python
info_asd = data_info_asd.describe()
info_asd.round()
```
